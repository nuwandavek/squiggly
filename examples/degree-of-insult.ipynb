```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("../")

import numpy as np
from datetime import datetime, timedelta
from squiggly import SquigglyPlot
```

```python
squigglyplot = SquigglyPlot(figsize=(20, 10))

# sf
xy = np.array(
    [[datetime(2015,1,1), .95],
    [datetime(2020,1,1), .8],
    [datetime(2020,6,1), 0.1],
    [datetime(2021,1,1), 0.1],
    [datetime(2022,7,1), 0.1],
    [datetime(2022,7,2), .5],
    [datetime(2022,11,29), .6],
    [datetime(2022,11,30), .99],
    [datetime(2023,3,19), .99]]
)
annotations = [
    [[datetime(2020, 1, 10), datetime(2020,4,1)], [.8, .9], [datetime(2020,2,1), .92], "Covid lockdown begins"],
    [[datetime(2022, 1, 30), datetime(2022,4,30)], [.2, .1], [datetime(2021,1,30), .22], "DallE, Stable diffusion release"],
    [[datetime(2022, 3, 30), datetime(2022,8,30)], [.7, .6], [datetime(2021,1,30), .72], "ChatGPT release"],
]
squigglyplot.draw_line(list(xy[:,0]), list(xy[:,1]), noise_strength=0.05, autocorr_perc=2, c="r", label='"Are you still in the Bay Area?"')
for anno in annotations:
    squigglyplot.draw_annotations(*anno, noise_strength=0.01)




squigglyplot.draw_grid([datetime(2015, 1, 1), datetime(2024,1,1)], [0, 1], legend=True)
squigglyplot.draw_title('Degree-of-Insult of a statement vs Time')
```
