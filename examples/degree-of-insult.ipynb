```python
squigglyplot = SquigglyPlot(figsize=(15, 15))
for line in lines:
    squigglyplot.draw_line(list(line['xy'][:,0]), list(line['xy'][:,1]), noise_strength=0.05, autocorr_perc=2,
                       c=line['color'], label=line['statement'])
    for anno in line['annotations']:
        squigglyplot.draw_annotations(*anno, noise_strength=0.01)



squigglyplot.draw_grid([datetime(2015, 1, 1), datetime(2024,1,1)], [0, 1], legend=True)
squigglyplot.draw_title('Degree-of-Insult of a statement vs Time \n (v0: SF Edition)')
squigglyplot.draw_annotations([], [], [datetime(2020,7,20), -.4], 'vivekaithal.co/squiggly_lines',
                              textbg=False, fontsize=30)
```
