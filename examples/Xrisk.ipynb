```python
squigglyplot = SquigglyPlot(figsize=(15, 10))
for line in lines:
    squigglyplot.draw_line(list(line['xy'][:,0]), list(line['xy'][:,1]), noise_strength=0.05, autocorr_perc=1,
                       c=line['color'], label=line['statement'])
    for anno in line['annotations']:
        squigglyplot.draw_annotations(*anno, noise_strength=0.01)



squigglyplot.draw_grid([datetime(1300, 1, 1), datetime(2024,1,1)], [0, 1], legend=True)
squigglyplot.draw_title('''Confidence of "Experts" that the 
"Next Big Technological Leap is going to literally kill all jobs" vs Time''')
squigglyplot.draw_annotations([], [], [datetime(1700,7,20), -.3], 'squigglylines.vivekaithal.co',
                              textbg=False, fontsize=30)
```
