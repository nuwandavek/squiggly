```python
import numpy as np
from datetime import datetime, timedelta

from squiggly import SquigglyPlot
```

```python
x = np.arange(0, 20, 0.1)
y = np.sin(x)
y2 = np.cos(x)


squigglyplot = SquigglyPlot(figsize=(20, 10))
squigglyplot.draw_line(x, y, c="r", label="sin wave")
squigglyplot.draw_line(x, y2, c="blue", label="cosine wave")
squigglyplot.draw_grid([0, 30], [-2, 2])
squigglyplot.draw_title('Degree-of-Insult of a statement vs Time')
squigglyplot.draw_annotations([7.5, 10], [0, 1], [9, 1], "hello world!")
squigglyplot.draw_annotations([11, 13], [-1, -1.5], [12, -1.6], "hello world!")
```

```python
x = np.arange(datetime(1985,7,1), datetime(2015,7,1), timedelta(days=10)).astype(datetime)
y = np.sin([p.timestamp() * 10 / x[0].timestamp() for p in x])
y2 = np.cos([p.timestamp() * 10 / x[0].timestamp() for p in x])


squigglyplot = SquigglyPlot(figsize=(20, 10))
squigglyplot.draw_line(x, y, c="r", label="sin wave")
squigglyplot.draw_line(x, y2, c="blue", label="cos wave")
squigglyplot.draw_grid([datetime(1990, 1, 1), datetime(2020,1,1)], [-2, 2])
squigglyplot.draw_title('Degree-of-Insult of a statement vs Time')
squigglyplot.draw_annotations([datetime(1997, 7, 1), datetime(2000,1,1)], [0, 1], [datetime(2000,1,1), 1], "hello world!")
```
